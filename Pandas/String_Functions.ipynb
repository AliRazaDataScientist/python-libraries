```python
import numpy as np
import pandas as pd
```

```python
rec = pd.read_csv('./file_data/bios.csv')
```

```python
rec['First_Name'] = rec['name'].str.split(' ').str[0]#use split function on string
```

```python
new_rec = rec.rename(columns={'name':'Full_Name'})#we are doing rename of columns
new_rec
```

```python
new_rec.query('First_Name == "Keith"')#find the string in the column
```

```python
new_rec.info()#get colums data
```

```python
new_rec['Born_Date'] = pd.to_datetime(new_rec['born_date'], errors='coerce')
#if the formate iss not proper so it through error so for controlling error we are using error = 'coerce'
```

```python
new_rec['Born_Year'] = new_rec['Born_Date'].dt.year# we are getting only year from the date function
new_rec.head()
```

```python
new_rec['Height_Category'] = new_rec['height_cm'].apply(lambda x:'Short' if x<165 else('Average' if x<180 else 'Tall' ))
#To run function on the specific column so we are using apply function
new_rec
```

```python
def categories_athlete(xyz):
    if xyz['weight_kg']<70 and xyz['height_cm']<175:
        return('Light Weight')
    elif xyz['height_cm']<185 or xyz['weight_kg']<85:
        return('Normal Weight')
    else:
        return('Heavy Weight')
new_rec['Category'] = new_rec.apply(categories_athlete, axis=1)
```
