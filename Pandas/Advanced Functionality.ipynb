```python
import pandas as pd
```

```python
bios = pd.read_csv('./file_data/bios.csv')
coffee = pd.read_csv('./file_data/Coffee.csv')
```

```python
coffee['Yesterday Revenue'] = coffee['Revenue'].shift(2)
```

```python
coffee['PCT Change'] = coffee['Revenue']/coffee['Yesterday Revenue']
```

```python
coffee['Cumulative_Revenue'] = coffee['Revenue'].cumsum()
```

```python
latte = coffee[coffee['Coffee Type'] == 'Latte'].copy()
```

```python
latte['3 Day'] = latte['Unit Sold'].rolling(3).sum()
```
